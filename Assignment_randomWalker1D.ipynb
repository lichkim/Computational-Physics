```python
from pylab import plot, show, xlabel, ylabel, title, hist
from random import random
```

```python
def randWalk1D(steps):
    xlist=list()    #결과값을 반환할 리스트
    pos=0    #현재 위치를 담을 변수
    for i in range(steps):
        rand=random()    #0~1 사이의 랜덤 숫자를 담을 변수
        if rand<0.5:
            pos+=1
        else:
            pos-=1
        xlist.append(pos)
    
    return xlist
```

```python
tryNum=int(input('Input steps : '))
result=randWalk1D(tryNum)
plot(range(1,tryNum+1),result)
xlabel('steps')
ylabel('position')
title('1D Random Walker Plot')
show()
```

```python
finalAbsPosList=list()    #원점으로부터의 거리를 담을 리스트. 1차원이라 절댓값이 곧 거리가 된다.
spots=int(input('Input the number of spots : '))
finalPos=list()           #최종 위치들을 담을 리스트
average=0
for i in range(spots):
    temp=randWalk1D(tryNum)
    finalAbsPosList.append(abs(temp[-1]))    #randWalk1D의 결과값의 마지막 값이 최종 위치가 된다.
    finalPos.append(temp[-1])
    average+=abs(temp[-1])/spots

hist(finalAbsPosList, bins=range(0,350,10))  #bins는 히스토그램 막대 크기를 결정하는 인자
xlabel('Distance')
ylabel('Frequency')
title(f'1D Random Walker MSD : {average}')
show()
```

```python
hist(finalPos,bins=list(range(-360,360+1,10)))
title('Final Position Distribution')
show()
```
