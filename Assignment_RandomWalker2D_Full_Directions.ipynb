```python
from pylab import plot, show, xlabel, ylabel, title, hist, scatter, hist2d
from random import random
from math import sqrt, sin, cos, pi
```

```python
def randomWalker2D(steps):
    xpos=0    #x의 위치를 담을 변수
    ypos=0    #y의 위치를 담을 변수
    xposlist=list()    #x의 모든 위치를 담을 리스트
    yposlist=list()    #y의 모든 위치를 담을 리스트
    
    for i in range(steps):
        rand=random()*2*pi    #0~2ㅠ 사이의 임의의 값 한 개가 나오게 된다.
        xpos+=cos(rand)    
        ypos+=sin(rand)    #x와 y에 각각 cost, sint 값을 더해주면 결과적으로 거리를 1만큼 이동한 것이 됨
        
        xposlist.append(xpos)
        yposlist.append(ypos)
        
    return [xposlist,yposlist]    #리스트를 담은 리스트를 반환함으로써 0번째 원소를 x축, 1번째 원소를 y축으로 사용 가능
```

```python
def distance(x,y):
    return sqrt(x**2+y**2)    #원점으로부터의 거리를 반환하는 함수
```

```python
steps=int(input('Input steps : '))
result=randomWalker2D(steps)
plot(result[0],result[1])    #result[0] : x축, result[1] : y축
plot(result[0][-1],result[1][-1],'ro')    #마지막 위치를 빨간 점으로 찍음
xlabel('Position of X')
ylabel('Position of Y')
title('Random Walker 2D with Full Directions')
show()
```

```python
spots=int(input('Input the number of spots : '))
final_distance=list()    #최종 위치의 거리를 담을 리스트
xfinal=list()    #최종 x의 위치를 담을 리스트
yfinal=list()    #최종 y의 위치를 담을 리스트
average=0
for i in range(spots):
    result=randomWalker2D(steps)
    xfinal.append(result[0][-1])    #마지막 원소만 담는다
    yfinal.append(result[1][-1])
    final_distance.append(distance(result[0][-1],result[1][-1]))
    average+=distance(result[0][-1],result[1][-1])**2/spots
hist(final_distance,bins=list(range(0,300,10)))    #bins는 히스토그램 막대의 크기를 결정하는 인자
xlabel('Distance of final position')
ylabel('Number of Spots')
title(f'Histogram of Final position : MSD={sqrt(average)}')
show()
```

```python
title('Final Distribution')
hist2d(xfinal,yfinal,bins=(list(range(-200,200,10)),list(range(-200,200,10))))    #hist2d는 2차원 히스토그램을 그리는 함수
show()
```
