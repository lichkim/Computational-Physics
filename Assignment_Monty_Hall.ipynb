```python
from random import randint
from pylab import plot, show, xlabel, ylabel, semilogx
```

```python
def monty_hall(steps):    #사회자가 문에 대한 정보를 알 때
    change=0
    stick=0
    change_list=list()
    stick_list=list()
    for i in range(steps):
        room=[False,False,False]
        room[randint(0,2)]=True
        player_choice=randint(0,2)
    
        if room[player_choice]:    #True라면 처음에 고른게 차, False라면 처음에 고른게 차가 아님
            stick+=1
        else:
            change+=1
        
        change_list.append(change/(i+1))
        stick_list.append(stick/(i+1))
    
    return [change_list,stick_list]
```

```python
def stupid_monty_hall(steps):    #사회자가 문에대한 정보가 없을 때
    change=0
    stick=0
    change_list=list()
    stick_list=list()
    
    for i in range(steps):
        room=[False]*3    #[False, False, False]와 같은 역할
        car=randint(0,2)
        room[car]=True
        
        while True:
            mc_choice=randint(0,2)
            if mc_choice==car:
                continue
            else:
                del room[mc_choice]
                break
        
        player_choice=randint(0,1)    #리스트에 원소가 하나 빠졌으니 0~1중 하나만 고르면 된다.
            
        if room[player_choice]:    #True라면 처음에 고른게 차, False라면 처음에 고른게 차가 아님
            stick+=1
        else:
            change+=1
        
        change_list.append(change/(i+1))
        stick_list.append(stick/(i+1))
    
    return [change_list,stick_list]
```

```python
steps=int(input('Input steps : '))
result=monty_hall(steps)
print(f'If you change your choice : {result[0][-1]}%')
print(f'If you stick to your choice : {result[1][-1]}%')
```

```python
semilogx(result[0],'r')
semilogx(result[1],'b')
xlabel('Times')
ylabel('Probability')
show()
```

```python
steps=int(input('Input steps : '))
result=stupid_monty_hall(steps)
print(f'If you change your choice : {result[0][-1]}%')
print(f'If you stick to your choice : {result[1][-1]}%')
```

```python
semilogx(result[0],'r')
semilogx(result[1],'b')
xlabel('Times')
ylabel('Probability')
show()
```
