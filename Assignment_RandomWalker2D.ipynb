```python
from pylab import plot, show, xlabel, ylabel, title, hist, scatter, hist2d
from random import randint
from math import sqrt
```

```python
def randomWalker2D(steps):
    xpos=0    #x의 위치를 담을 변수
    ypos=0    #y의 위치를 담을 변수
    xposlist=list()    #x의 모든 위치를 담을 리스트
    yposlist=list()    #y의 모든 위치를 담을 리스트
    
    for i in range(steps):
        rand=randint(1,4)    #randint : 1과 4 사이의(4 포함) 랜덤한 정수 하나를 반환함
        if rand==1:
            xpos+=1
        elif rand==2:
            xpos-=1
        elif rand==3:
            ypos+=1
        else:
            ypos-=1
        
        xposlist.append(xpos)
        yposlist.append(ypos)
        
    return [xposlist,yposlist]  #리스트가 담긴 리스트를 반환함으로써 리스트의 0번째 원소, 1번째 원소를 각각 x축 y축으로 사용 가능
```

```python
def distance(x,y):
    return sqrt(x**2+y**2)    #원점으로부터의 거리를 반환하는 함수
```

```python
steps=int(input('Input Steps : '))
result=randomWalker2D(steps)
plot(result[0],result[1])    #result[0] : x축, result[1] : y축
plot(result[0][-1],result[1][-1],'ro')    #마지막 위치를 빨간 점으로 표시
xlabel('Position of X')
ylabel('Position of Y')
title('Random Walker 2D')
show()
```

```python
spots=int(input('Input the number of spots : '))
final_distance=list() #최종 위치의 거리를 담을 리스트
final_xpos=list()    #최종 x의 위치
final_ypos=list()    #최종 y의 위치
average=0
for i in range(spots):
    result=randomWalker2D(steps)
    final_distance.append(distance(result[0][-1],result[1][-1]))
    final_xpos.append(result[0][-1])
    final_ypos.append(result[1][-1])
    average+=distance(result[0][-1],result[1][-1])**2/spots
    
hist(final_distance,bins=range(0,360,10)) #bins는 히스토그램의 크기를 결정하는 인자
xlabel('Distance')
ylabel('Number of Spots')
title(f'Final distance of Random Walker 2D / MSD : {sqrt(average)}')
show()
```

```python
hist2d(final_xpos,final_ypos,bins=(list(range(-200,200,10)),list(range(-200,200,10)))) #array_like 인자를 2개를 줄 수 있음(matplolib 라이브러리 참조)
title('Final Position Distribution')
show()
```
